```python
from description_embedder import DescriptionEmbedder
from tags_embedder import TagsEmbedder
import pandas as pd
```

```python
# Load data
data = pd.read_csv('../data/clean_data.csv', sep=";").head(20)
data
```

```python
desc_embd = DescriptionEmbedder()
```

```python
description_embeddings = desc_embd.create_embeddings(data)
description_embeddings
```

```python
from sentence_transformers import SentenceTransformer
from nltk.tokenize import sent_tokenize
model = SentenceTransformer('all-mpnet-base-v2')
data['sentences'] = data['About the game'].apply(sent_tokenize)
data['embeddings'] = data['sentences'].apply(lambda x: model.encode(x).mean(axis=0))
```

```python
for i, entries in enumerate(data['embeddings'].values):
    print((description_embeddings[i] == entries).all())
```

```python
tags_embd = TagsEmbedder()
tags_embeddings = tags_embd.create_embeddings(data)
tags_embeddings
```

```python
import gensim.downloader as api
import numpy as np
w2v = api.load('word2vec-google-news-300')

data['tags_token'] = data['Tags'].apply(lambda x: x.split(','))
data['tags_embeddings'] = data['tags_token'].apply(lambda x: np.mean([w2v[word] for word in x if word in w2v], axis=0))
data['tags_embeddings']
```

```python
for i, entries in enumerate(data['tags_embeddings'].values):
    print((tags_embeddings[i] == entries).all())
```
